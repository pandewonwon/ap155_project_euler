```python
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal
from fractions import Fraction
np.seterr(divide='ignore', invalid='ignore')
```

```python
#We will study the following function:
#f(x)=(1-cosx)/(x^2)
#(a) Start by plotting the funciton, using a grid of the form x=0.1(i) for i=1,2,...,100. This should give you some idea of the values you 
#should expect for f(x) at small x.
#(b) Verify your previous hunch by taking the limit x-->0 and using L'Hopital's rule.
#(c) Now, see what value you find for f(x) when x=1.2x10^(-8). Does this make sense, even quantitatively?
#(d) Use a trigonometric identity to avoid the cancellation. Evaluate the new function at x=1.2x10^(-8) and compare with your analytical answer
#for x-->0
```

```python
#plotting
fig,ax = plt.subplots()

for i in range(1,101):
    x=np.float128(0.1)*i
    t=np.linspace(0,x,500)
    y=(1-np.cos(t))/(t**2)
    
    ax.plot(t,y)

x=1.2*(10**(-8))
y_at_some_point = (1-np.cos(x))/(x**2)
ax.plot(x,y_at_some_point,"ro")
print("Using L'Hopital's rule, the limit of the function as x-->0 is 0.5. Based on the graph, the function does approach 1/2 or 0.5 as x-->0.")
print("y at some point=", y_at_some_point)
print("y at some point does not make sense because the limit of the function as x-->0 should be 0.5. Quantitatively and qualitatively, we witness a different behavior when the function is evaluated at x=1.2x10^(-8).")
```

```python
#plotting
fig,ax = plt.subplots()

for i in range(1,101):
    x=np.float128(0.1)*i
    t=np.linspace(0,x,500)
    y=(1/2)*((np.sin(t/2))/(t/2))**2 #Equivalent function using trigonometric identities
    
    ax.plot(t,y)

x=1.2*(10**(-8))
y_at_some_point = (1/2)*((np.sin(x/2))/(x/2))**2
ax.plot(x,y_at_some_point,"ro")

print("y at some point=",y_at_some_point)
print("Using the equivalent function, the limit as x-->0 and the evaluation at x=1.2x10^(-8) are both equal to 0.5. This shows that the equivalent function is more stable and can avoid cancellations. The original function is unstable or undefined at x=0 due to the denominator.")
```
