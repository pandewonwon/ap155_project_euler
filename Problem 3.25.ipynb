```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots()

x=np.linspace(0,2*np.pi,128)

def function_f(x):
    f_x=2+5*np.sin(x)+0.1*np.sin(30*x)
    return f_x

def function_g(x):
    g_x=2+5*np.sin(x)
    return g_x

ax.plot(x,function_f(x),label="f(x) with strong oscillations")
ax.plot(x,function_g(x),"--", label="g(x)without strong oscillations")
ax.set(
    xlabel="x",
    ylabel="y")
ax.legend()
plt.savefig("3.25 functions.png")
```

```python
print("Table of Values x and corresponding f(x)")
print(x,function_f(x))
```

```python
fig,ax=plt.subplots()

#analytically derived f'(x)
f_prime=5*np.cos(x)+3*np.cos(30*x)

#analytically derived g'(x)
g_prime=5*np.cos(x)

#forward-difference approx. for f'(x)
def fd_fprime(func,x,h):
    h=x[1]-x[0] #stepsize. h is the distance between points
    fd=(function_f(x+h)-function_f(x))/(h)
    return fd    

fprime_forward=fd_fprime(function_f,x,h)

def fd_second(func,x,h):
    h=x[1]-x[0] #stepsize. h is the distance between points
    second=(function_f(x+2*h)-function_f(x))/(2*h)
    return second

fprime_second=fd_second(function_f,x,h)

ax.plot(x,f_prime,label="analytical f'(x)")
ax.plot(x,g_prime,label="analytical g'(x)")
ax.plot(x,fprime_forward,"--",label="f'(x) Forward Difference")
ax.plot(x,fprime_second,label="f'(x) using twice removed grid points")
ax.set(
    xlabel="x",
    ylabel="y")
ax.legend()
plt.savefig("3.25 function derivatives.png")
```
