```python
Problem 1:
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3,5,6,and 9. The sum of these multiples if 23. 
Find the sum of all multiples of 3 or 5 below 1000.
```

```python
multiples_of_3=[i for i in range(1,1000) if i%3==0]
print(multiples_of_3)
```

```python
multiples_of_5=[i for i in range(1,1000) if i%5==0]
print(multiples_of_5)
```

```python
multiples_of_3_ANDOR_5 = multiples_of_3 + multiples_of_5
print(multiples_of_3_OR_5)
```

```python
multiples_of_3_OR_5 = set(multiples_of_3_ANDOR_5)
print(multiples_of_3_OR_5)
```

```python
list = []
for i in multiples_of_3_OR_5:
    list.append(i)

print(list)
```

```python
total = 0 
for item in list:
    total += item

print(total)
```
