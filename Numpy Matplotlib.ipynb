```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#example
t = np.arange(0,10,0.05)
freq = 1.0

y = np.sin(2*np.pi*freq*t)

fig,ax = plt.subplots()

shifted_y = np.sin(2*np.pi*freq*t-2.0)

ax.plot(t,y, label="not shifted")
ax.plot(t,shifted_y, label="shifted")
ax.legend()
```

```python
#Create two sinusoid functions with different frequencies f_1 and f_2.
#Plot the element-wise product of these two signals. Check how the difference in frequencies and shifting phases affect the resulting signal.
```

```python
x = np.arange(0,10,0.05)

freq_1 = 1.0
freq_2 = 1.25

sin1 = np.sin(2*np.pi*freq_1*x)
sin2 = np.sin(2*np.pi*freq_2*x)

product = sin1*sin2
```

```python
fig,ax = plt.subplots()
ax.plot(x,sin1,"--", label="function 1")
ax.plot(x,sin2,"--",label="function 2")

ax.legend()
```

```python
fig,ax = plt.subplots()
ax.plot(x,product, label="product")

ax.legend()
```

```python
#Make a quarter circle using np.arange with radius 1 
#Draw uniformly random numbers that represent random points(x_i,y_i) in the figure, then create a scaltter plot of all the points that fall within the quarter circle.
#The figure is limited to cover 0 to 1 in x and y.
```

```python
fig,ax = plt.subplots()

t = np.linspace(0,1,100)
x = np.sqrt(1-t**2)

inside = 0
outside = 0

for i in range(1000):
    x_i = np.random.random()
    y_i = np.random.random()
    if x_i**2+y_i**2<1:
        ax.scatter(x_i,y_i)
        inside += 1
    else:
        outside += 1

ratio = inside/outside

print(inside,"inside")
print(outside,"outside")
print(ratio,"ratio")
ax.plot(t,x,label="quarter circle")
ax.legend()
```

```python
#or alternatively, a more efficient way:
fig,ax = plt.subplots()

N = 10000
x_i = np.random.random(size = N)
y_i = np.random.random(size = N)

is_inside = x_i**2+y_i**2 < 1

ax.scatter(x_i[is_inside],y_i[is_inside])
print(4*np.sum(is_inside)/N)
```

```python
fig,ax = plt.subplots()

lognormal = np.random.lognormal(5,0.1,size = N) #[mean,sigma,size]

_ = ax.hist(lognormal, bins=np.linspace(100,300,1000), density=True)
```
