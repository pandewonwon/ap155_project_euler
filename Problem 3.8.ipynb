```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This problem studies the second derivative of the following function:
#f(x)=(1-cos(x))/x^2
#Take x=0.004

#a. start by analytically evaluating the second derivative, and plugging in x
#b. Give h the  values 10^-1 to 10^-6 and produce a log-log plot of the absolute error in the first central-difference approximation to the second derivative.
#c. Introduce a new set of points to the previous plot, this time, evaluating the first central-difference approximation to the second derivative not of f(x), but of the analytically rewritten
#version you arrived at in Problem 2.8.
```

```python
def f_func(x):
    y=(1-np.cos(x))/(x**2)
    return y
    
def central_diff(f_func,x,h): 
    doubleprime_cd=((f_func(x+h))-(2*f_func(x))+(f_func(x-h)))/(h**2)
    return doubleprime_cd
```

```python
x=0.004

f_doubleprime = (x**2*np.cos(x)-4*x*np.sin(x)+6-6*np.cos(x))/(x**4) #analytically derived
print(f_doubleprime)
```

```python
hs=[10**(-i) for i in range(1,7)]
hs=np.logspace(-7,-1,200)
ans=[abs(central_diff(f_func,x,h)-f_doubleprime) for h in hs]

fig,ax=plt.subplots()
ax.loglog(hs,ans)
ax.set_title("Absolute Error (Central Difference) of Unstable function")

plt.savefig("Figure.png")
```

```python
def equiv_func(x):
    y=(1/2)*((np.sin(x/2))/(x/2))**2
    return y
    
def central_diff(equiv_func,x,h): 
    doubleprime_cd=((equiv_func(x+h))-(2*equiv_func(x))+(equiv_func(x-h)))/(h**2)
    return doubleprime_cd
```

```python
hs=[10**(-i) for i in range(1,7)]
hs=np.logspace(-7,-1,200)
ans=[abs(central_diff(equiv_func,x,h)-f_doubleprime) for h in hs]

fig,ax=plt.subplots()
ax.loglog(hs,ans)
ax.set_title("Absolute Error(Central Difference) of Stable Function")
```
